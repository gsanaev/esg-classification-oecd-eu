```python
import pandas as pd

df = pd.read_csv("..//data/raw/oecd_ghg_raw_2015_2025.csv")

# Focus on one year to test
sample = df[df["TIME_PERIOD"] == 2019]
print(sample.groupby("REF_AREA")["OBS_VALUE"].mean().head())

# Now try a mock region aggregation
region_map = {
    "DEU": "Europe", "FRA": "Europe", "ITA": "Europe",
    "USA": "North America", "CAN": "North America",
    # â€¦ etc.
}

sample["Region"] = sample["REF_AREA"].map(region_map)
agg_wrong = sample.groupby("Region")["OBS_VALUE"].sum()
agg_right = sample.groupby("Region")["OBS_VALUE"].mean()

print("\nWRONG aggregation (sum):\n", agg_wrong)
print("\nCORRECT aggregation (mean):\n", agg_right)
```
